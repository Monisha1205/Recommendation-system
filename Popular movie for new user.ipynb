```python
import pandas as pd # import pandas library
```

```python
# import data
path_to_dataset = 'ml-100k/u.data'
columns = ['user_id', 'item_id', 'rating', 'timestamp']
ratings = pd.read_csv(path_to_dataset, sep='\t', names=columns)
```

```python
# import item file
path_to_movies = 'ml-100k/u.item'
movie_columns = ['item_id', 'title', 'release_date', 'video_release_date', 'IMDb_url',
                 'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime',
                 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
                 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies = pd.read_csv(path_to_movies, sep='|', encoding='latin-1', names=movie_columns)
```

```python
# merge ratings and movies
movie_ratings = pd.merge(ratings, movies[['item_id', 'title', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']], on='item_id')
movie_ratings
```

```python
categories = ['Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

top_movies_by_category = {}

for category in categories:
    category_movies = movie_ratings[movie_ratings[category] == 1]
    top_movies = category_movies.groupby('title')['rating'].mean().sort_values(ascending=False).head(5)
    top_movies_by_category[category] = top_movies
```

```python
for category, top_movies in top_movies_by_category.items():
    print(f"\nTop 5 Movies in {category}:\n")
    print(top_movies)
```
